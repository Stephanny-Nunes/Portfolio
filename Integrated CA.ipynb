## 1 Introduction

<p style="text-align:justify"><span style="font-size:1.05em;">In face of climate change and global warming, it is getting harder to understand rain pattern around the globe.
```python
# Adding Measures of Variability to the dataframe: Standard Deviation, Variance and IQR

CTM_num['Std'] = rain_df[num].std()
CTM_num['Variance'] = rain_df[num].var()
CTM_num['IQR'] = (rain_df[num].quantile(0.75)) - (rain_df[num].quantile(0.25))
```

## 4 Treating the values

<p style="text-align:justify"><span style="font-size:1.05em;">Before performing visualisation of the variables, the missing datapoints will be treated, a few varibles will be encoded and the variable date will be converted.</span></p>

### 4.1 Handling missing values

<p style="text-align:justify"><span style="font-size:1.05em;">To eliminate missing data in numerical variables, a for loop was performed to take all the numerical columns and to fill the missing values with the mean. It was decided that the missing values in the variables <i><b>RainToday</b></i> and <i><b>RainTomorrow</b></i> would be dropped, because the number of missing data is really low. The other categorical variables were filled with the mode, as it is in the dataframe <i>CTM_cat</i>. </span></p>

```python
# Performing a for loop to take all numerical features and fill the NaN values with the mean

for feat in rain_df.columns:
    if rain_df[feat].dtypes != 'object':
        rain_df[feat] = rain_df[feat].fillna(rain_df[feat].mean())
```

```python
# Filling a few categorical variables with the mode, according to the dataframe created before

rain_df['WindGustDir'] = rain_df['WindGustDir'].fillna('W')
rain_df['WindDir9am'] = rain_df['WindDir9am'].fillna('N')
rain_df['WindDir3pm'] = rain_df['WindDir3pm'].fillna('SE')
```

```python
# Droping the rest of the null values (RainToday and RainTomorrow)

rain_df = rain_df.dropna()
```

```python
# Checking if there is still missing values

rain_df.isnull().sum()
```

### 4.2 Encoding Categorical Variables

<p style="text-align:justify"><span style="font-size:1.05em;">The variables <i><b>RainToday</b></i> and <i><b>RainTomorrow</b></i> were enconded now because they have to be seen in the correlation heatmap.</span></p>

```python
enc = OrdinalEncoder(dtype = np.dtype(int))        # Creating varible to store the Ordinal Encoder method
def encode(col):                                   # Creating function encode to apply in the columns selected
    y = rain_df[col].values.reshape(-1,1)          # Storing columns reshaped in y
    enc.fit(y)                                     # Fitting y with Ordinal Encoder (enc)
    rain_df[col] = enc.transform(rain_df[col].values.reshape(-1,1))   # Transforming the values in the column
```

```python
encode('RainToday')            # Applying encode function on column RainToday
encode('RainTomorrow')         # Applying encode function on column RainTomorrow
```

## 6 Linear Regression for Statistics

<p style="text-align:justify"><span style="font-size:1.05em;">As informed previously, the heatmap showed how strong is the correlation among variables. The variables <i><b>MaxTemp</b></i> and <i><b>Temp9am</b></i> were chosen. Applying the correlation function, the result is <b>0.88</b>, which means there is a strong correlation between these two variables. It is clear, by looking at the scatterplot below, that there is a pattern between the variables.</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">To understand this pattern, it was performed a n.polyfit with degree one to get the coeficients of the linear equation that reflects the correlation between <i><b>MaxTemp</b></i> and <i><b>Temp9am</b></i>.</span></p>

```python
# Correlation matrix for variables chosen
rain_df[['MaxTemp', 'Temp9am']].corr(method='pearson')
```

```python
# Use variables chosen to plot scatterplot

sns.scatterplot(data=rain_df, x="MaxTemp", y="Temp9am");
```

```python
rain_fit = np.polyfit(rain_df.MaxTemp, rain_df.Temp9am, 1)

rain_fit
```

<p style="text-align:justify"><span style="font-size:1.05em;">After performing that, the coefficients calculated were <b>a=0.80</b> and <b>b=-1.70</b>. Using these values, it is possible to create the Linear Regression Model: <b>y = 0.80x - 1.70</b>.</span></p>
<p style="text-align:justify"><span style="font-size:1.05em;">MaxTemp: x</span></p>
<p style="text-align:justify"><span style="font-size:1.05em;">Temp9am: y</span></p>

```python
# Plotting the scatter plot
rain_df.plot(kind='scatter', x='MaxTemp', y='Temp9am', color='blue', alpha=0.5)

# Plotting Regression Line
plt.plot(rain_df.Temp9am, rain_fit[0] * rain_df.Temp9am + rain_fit[1], color='red', linewidth=3)

# Plotting Regression Equation
plt.text(25, 0,'y={:.2f}+{:.2f}*x'.format(rain_fit[1],  rain_fit[0]), color='red', fontsize=14)

# legend, title and labels.
plt.legend(labels=['Rain Regresion Line', 'Temp9am', 'MaxTemp'])
plt.title('Relationship between Temp9am and MaxTemp \n', fontsize=14, fontweight='bold')
plt.xlabel('Temp9am', fontsize=12)
plt.ylabel('MaxTemp', fontsize=12);
```

<p style="text-align:justify"><span style="font-size:1.05em;">The method used above gave us the coefficients and the Linear Model. It is possible to perform prediction using it, but there is a library on python to do that. It will be performed below the Linear Regression using Scikit Learn.</span></p>

## 9 Reinforcement Learning

<p style="text-align:justify"><span style="font-size:1.05em;">Reinforcement Learning is the branch of Machine Learning where the model learns based on a reward/penalty method. "It's a feedback-based machine learning method in which the AI agent learns to behave in an environment by taking actions and seeing those actions' results" (Joby, 2021). If the machine performs the way it is expected to, it is rewarded, otherwise it is penalised. The agent performs an action in the determined environment, then the agent recieves a reward, positive or negative. The state (the situation or position the agent is after performing the action)  helps the agent to determine the policy (strategy the agent uses to procced based on the current state) and the cycle repeats.</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">Reinforcement Learning is highly used in real-world fields, such as: Industry automation, Trading and Finance and Robotics manipulation.</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">There are two types of Reinforcement Learning: Passive and Active. In the Passive RL the agent executes a fixed policy that was stablished for it, while in Active RL the agent acts and learns an optimal policy.</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">Passive Learning can be done in three different ways. First one is called Direct Utility Estimation. In this method "the agent executes a sequence of trials or runs (sequences of states-actions transitions that continue until the agent reaches the terminal state). Each trial gives a sample value and the agent estimates the utility based on the samples values" (Bahtt, 2018) </span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">Second one is called Adaptive Dynamic Programming (ADP). This method "runs trials to learn the model of the environment by estimating the utility of a state as a sum of reward for being in that state and the expected discounted reward of being in the next state" (Bahtt, 2018).</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">The third one is called Temporal Difference Learning. This method "does not require the agent to learn the transition model. The update occurs between successive states and agent only updates states that are directly affected" (Bahtt, 2018).</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">An exemple of active RL is the Speech Recognition system. "Certain software applications can convert live voice and recorded speech into a text file. The speech can be segmented by intensities on time-frequency bands as well" (Guerrouj, 2020).</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">Real-life examples of Reinforcement learning that are being used right now: "Shell is using deep-learning algorithms that are trained from historical drilling data, as well as data from simulations, to steer the gas drills as they move through a subsurface. The DRL technology also includes the mechanical data from the drill bit, such as pressure and bit temperature, as well as seismic survey data relevant to the subsurface. As a result, the human operator of the drilling machine has a better understanding of the environment they’re working in, which leads to quicker results, and less wear and tear – or damage – to expensive drilling machinery" (Get Smarter, 2019).</span></p>

<p style="text-align:justify"><span style="font-size:1.05em;">Another example of Active Reinforcement learning is the self-driving cars. According to Mwiti (2021) "parking can be achieved by learning automatic parking policies. Lane changing can be achieved using Q-Learning while overtaking can be implemented by learning an overtaking policy while avoiding collision and maintaining a steady speed thereafter". Q-Learning is a type of Active RL method that does not require the agent to learn the transitional model, instead learns Q-value functions Q(s, a).</span></p>
